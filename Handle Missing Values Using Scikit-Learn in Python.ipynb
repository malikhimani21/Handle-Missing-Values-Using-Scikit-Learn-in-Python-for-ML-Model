```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
"""
SimpleImputer is a scikit-learn class which is helpful in
handling the missing data in the predictive model dataset.
"""
```

```python
#Load data

train = pd.read_csv("C:\\Users\\Felix ITs 01\\Desktop\\train.csv")

test = pd.read_csv("C:\\Users\\Felix ITs 01\\Desktop\\test.csv")

print("Train df = ",train.shape)

print("Test df = ",test.shape)
```

```python
# Drop target column SalePrice

X_train = train.drop(columns="SalePrice")

y_train = train["SalePrice"]

print("X_train df = ",X_train.shape)

print("y_train df = ",y_train.shape)
```

```python
# Handle Numerical Missing Value
# Get All int and float columns

num_vars = X_train.select_dtypes(include=["int64","float64"]).columns
num_vars
```

```python
# Check null values

X_train.isnull()
```

```python
# Check All columns

pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)

X_train.isnull().sum()
```

```python
X_train.isnull().sum().sum()
```

```python
# Define strategy for handling missing values
# create class object

imputer_mean = SimpleImputer(strategy='mean')
```

```python
# Fit method()

imputer_mean.fit(X_train[num_vars])
```

```python
"""
fit function in ML, is used for training of model using data examples.
Fit function adjusts weights according to data values so that better accuracy can be achieved.
After training, the model can be used for predictions.
"""
```

```python
# You’ll get mean values for all columns

imputer_mean.statistics_
```

```python
"""
transform() : parameters generated from fit() method
The transform method is transforming all the features using the respective mean and variance.
"""
```

```python
imputer_mean.transform(X_train[num_vars])
```

```python
X_train[num_vars] = imputer_mean.transform(X_train[num_vars])

test[num_vars] = imputer_mean.transform(test[num_vars])
```

```python
"""
Here we got array of data.
Missing value filled with mean value.
We didn’t get column name here. Here data is in 2D array form.
"""
```

```python
# Handle String Missing data (Categorical variable )
# Here O means object data

cat_vars = X_train.select_dtypes(include=["O"]).columns

cat_vars
```

```python
X_train[cat_vars].isnull().sum()
```

```python
# Define strategy for cat data

imputer_mode = SimpleImputer(strategy='most_frequent')
imputer_mode
```

```python
imputer_mode.statistics_
```
